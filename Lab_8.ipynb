```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hist_eq():
    pname = "image.png"
    im = Image.open(pname).convert('L')
    pix_val = list(im.getdata())
    plt.hist(pix_val, normed =True, bins = 255)
    plt.show()

    gray_levels = {i:0 for i in range(0,256)} #initializing keys with 0 from 0 to 255
    pix_list = list(pix_val)
    for pix in pix_val:
        gray_levels[pix] +=1 #Finding no of pixels of each gray level in picture

    M,N = im.size
    size = M*N
    pdf = {i:(float(gray_levels[i])/size) for i in range(0,256)}

    cdf = {i:pdf[i] for i in range(0,256)}
    for i in range(1,256):
        cdf[i] = pdf[i]+cdf[i-1] 

    s = {i:int(round(255*cdf[i])) for i in range(0,256)}

    width, height = im.size
    im2 = Image.new('L',im.size)# make a new image in gray scale
    for x in range(0,width-1):
        for y in range(0, height-1):
            im2.putpixel((x,y), s[im.getpixel((x,y))])

    #New Image
    pix_val = list(im2.getdata())
    plt.hist(pix_val, normed =True, bins = 255)
    plt.show()
    im2.show()
```

```python
hist_eq()
```
