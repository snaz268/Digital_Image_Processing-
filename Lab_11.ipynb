```python
from PIL import Image
im =  Image.open("Untitled.png")
im
```

```python
import cv2 
img = cv2.imread('Untitled.png',0)
ret1,th1 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret1)
print(th1)
img2 = Image.fromarray(th1)
img2
```

```python
import numpy as np
kernel1 = np.ones((3,2),np.uint8)
dilation1 = cv2.dilate(th1, kernel1, iterations=11) 
print('Dilation Iteration 1')
display(Image.fromarray(dilation1))

kernel2 = np.ones((3,3),np.uint8)
erosion1 = cv2.erode(dilation1,kernel2,iterations = 5)
print('Erosion Iteration 1')
display(Image.fromarray(erosion1))
```

```python
import matplotlib.pyplot as plt
r, l = cv2.connectedComponents(erosion1)
plt.imshow(l, cmap = "nipy_spectral")
```
