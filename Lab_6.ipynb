# DIP Lab 6: Image Enhancement

```python
#Original Image
from PIL import Image, ImageOps
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np

response = requests.get("https://outpostfortcollins.com/wp-content/uploads/sites/34/2017/06/the-outpost-cottages-grayscale.jpg")
im = Image.open(BytesIO(response.content)).convert('L')
pix_val = list(im.getdata())
plt.hist(pix_val, normed =True, bins = 255)
plt.show()
im
```

```python
#Histogram Equalization
im_eq = ImageOps.equalize(im)
im_eq = im_eq.convert("L")
pix_val = list(im_eq.getdata())
plt.hist(pix_val, normed =True, bins = 255)
plt.show()
im_eq
```

```python
#Contrast Stretching
a = 0
b = 255
c = int(np.amin(im))
d = int(np.amax(im))
```

```python
width, height = im.size
im2 = Image.new('L', im.size) # make a new image in gray scale
for x in range(0,width):
    for y in range(0,height):
        r = im.getpixel((x,y))
        s = abs(((r-c)*((b-a)/(d-c)))+a)
        im2.putpixel((x,y),s)
im2
```

## Analysis and Report:
Two methods were used to improve the contrast of a washed out image; histogram equaliztion and contrast stretching. 

### Histogram Equaliztion:
Histogram Equaliztion has certainly improved the contrast of the image but it has also given rise to false contours in the resultant image.

### Contrast Stretching:
Using this method, the resultant image's contrast is not as good as with histogram equalization. However, there is no visible false contouring.
